```julia
using Plots
using LinearAlgebra
```

```julia
f(x)=x[1]+x[2]-x[4]^2
```

```julia
#example of Superposition
a=[-2,0,1,-3]
f(x)=a'*x

x = [2, 2, -1, 1]; y = [0, 1, -1, 0];
alpha = 1.5; beta = -3.7;

lhs=f(alpha*x+beta*y)
rhs=alpha*f(x)+beta*f(y)
@assert lhs==rhs
```

```julia
#average function example
avg(x) = sum(x)/length(x)
x=[1,-3,2,-1]
avg(x)
```

```julia
#taylot approximation example
f(x)=x[1]+exp(x[2]-x[1]);
grad_f(z)=[1-exp(z[2]-z[1]),exp(z[2]-z[1])];
z=[1,2]
f_hat(x)=f(z)+grad_f(z)'*(x-z)
f([1,2]), f_hat([1,2])
```

```julia
#norm
x=[2,-1,2]
norm(x)
# RMS value
rms(x)=norm(x)/sqrt(length(x));
t=0:0.01:1;
x=cos.(8*t)-2*sin.(11*t)
avg(x)
rms(x)
plot(t,x)
plot!(t,avg(x)*ones(length(x)),color=:orange)
#plot!(t,(avg(x)+rms(x))*ones(length(x)),color=:green)
#plot!(t,(avg(x)-rms(x))*ones(length(x)),color=:red)
plot!(t,rms(x)*ones(length(x)),color=:pink)
plot!(legend=false)
```
