```julia
using Plots

# Parameters
g = 9.81  # acceleration due to gravity (m/s^2)
L = 1.0   # length of the pendulum (m)

# Solve using RK45 integrator
h = 0.001  # Initial step size

function f(u)
    θ, ω = u
    θ_dot = ω
    ω_dot = -(g / L) * sin(θ)
    return [θ_dot, ω_dot]
end

function rk45_step( u)
   f1=f(u)
   f2=f(u+0.5*h*f1)
   f3=f(u+0.5*h*f2)
   f4=f(u+h*f3)
    
   return u+(h/60)*(f1+2*f2+2*f3+f4)
    
end

function rk45_solve(t0, u0, tf)
    t = t0
    u = u0
    ts = [t]
    us = [u0]
    
    while t < tf
        u_new = rk45_step(u)
        t += h
        u = u_new
        push!(ts, t)
        push!(us, u)
    end
    
    return ts, us
end



# Initial conditions
θ₀ = π / 2  # initial angular displacement
ω₀ = 0.0    # initial angular velocity
u₀ = [θ₀, ω₀]

# Time span
t0 = 0.0
tf = 10.0

t_vals, u_vals = rk45_solve(t0, u₀, tf, h_init)

# Extract angle values
θ_vals = [u[1] for u in u_vals]

# Plot the results
plot(t_vals, θ_vals, label="Angle", xlabel="Time", ylabel="Angle", legend=true)
```
