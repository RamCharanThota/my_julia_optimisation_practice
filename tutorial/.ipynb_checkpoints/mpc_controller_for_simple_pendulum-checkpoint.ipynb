```julia
import Pkg
Pkg.add("ControlSystems")
using ControlSystemBase
using LinearAlgebra

using Plots
```

```julia

# Parameters
g = 9.81  # acceleration due to gravity (m/s^2)
L = 1.0   # length of the pendulum (m)
M =1.0  # mass of the pendulum 

# Solve using RK45 integrator
h = 0.001  # Initial step size

function f(u,τ)
    θ, ω = u
    θ_dot = ω
    ω_dot = -(g / L) * sin(θ)+(τ/(M*L^2))
    return [θ_dot, ω_dot]
end

function rk45_step(u,τ)
   f1=f(u,τ)
   f2=f(u+0.5*h*f1,τ)
   f3=f(u+0.5*h*f2,τ)
   f4=f(u+h*f3,τ)
    
   return u+(h/6)*(f1+2*f2+2*f3+f4)
    
end


function torque_function(t)
    return 0
end

function rk45_solve(t0, u0, tf)
    t = t0
    u = u0
    ts = [t]
    us = [u0]
    
    while t < tf
        u_new = rk45_step(u,torque_function(t))
        t += h
        u = u_new
        push!(ts, t)
        push!(us, u)
    end
    
    return ts, us
end



# Initial conditions
θ₀ = pi-0.001  # initial angular displacement
ω₀ = 0.0    # initial angular velocity
u₀ = [θ₀, ω₀]

# Time span
t0 = 0.0
tf = 100.0

t_vals, u_vals = rk45_solve(t0, u₀, tf)

# Extract angle values
θ_vals = [u[1] for u in u_vals].*180/pi
w_vals=[u[2] for u in u_vals]*180/pi
# Plot the results
plot(t_vals, θ_vals, label="Angle", xlabel="Time", ylabel="Angle", legend=true)
plot!(t_vals, w_vals, label="omega", xlabel="Time", ylabel="omega", legend=true)

```
