```julia
using Plots
using LinearAlgebra
```

```julia
#plot convex function
f(x)=x^2
#generate x valus
x= -5:0.1:5
# calculate corresponding y values
y=f.(x)
plot(x, y, label="f(x) = x^2", xlabel="x", ylabel="f(x)", legend=true)

```

```julia
# plot strong convex
# example of strong convexity
m=1 # parameter should be greater than zero
g(x)= f.(x)+(m/2)*norm(x)

#generate x valus
x= -5:0.1:5
# calculate corresponding y values
z=g.(x)
plot(x, y, label="f(x)", xlabel="x", ylabel="f(x)", legend=true)
plot!(x, z, label="g(x)1", xlabel="x", ylabel="f(x)", legend=true)
m=2 # parameter should be greater than zero
g(x)= f.(x)+(m/2)*norm(x)
z2=g.(x)

plot!(x, z2, label="g(x)2", xlabel="x", ylabel="f(x)", legend=true)

```
